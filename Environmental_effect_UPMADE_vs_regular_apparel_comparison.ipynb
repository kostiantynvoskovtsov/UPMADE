```python
#import all the libraries neede
import pandas as pd
import numpy as np
import plotly.express as px

#create a dictionae with values
df= pd.DataFrame(
    {'Metric name':['Climate change & Global Warming', 'Energy', 'Land use', 'Water use', 'Ecotoxicity', 'Acidification', 'Human health(oxide formation)'],
     'Metric unit':['kg, CO2 Eq', 'MJ(megajoules)', 'percentage', 'm3, world Eq', 'comparative toxic units(CTUe)', 'mol, H+-Eq', 'kg, NMVOC-Eq'],
     'Upmade t-shirt':[0.6978,9.9050,0.7643, 0.0772, 1.7975, 0.0026, 0.0022],
     'Regular t-shirt':[8.8078, 112.5722, 101.5165, 60.1677, 111.4971, 0.0584, 0.029]})
display(df)

#convert data frame from wide to long format
df_melted = df.melt(id_vars = ['Metric name', 'Metric unit'],
                    value_vars = ['Upmade t-shirt', 'Regular t-shirt'],
                    var_name = 'T-shirt type',
                    value_name = 'Value')
#display(df_melted)

#visulaise results
color_dict = {'Upmade t-shirt':'#000BEC','Regular t-shirt':"#373900"}
fig=px.bar(df_melted, x='Metric name', y='Value',
           color='T-shirt type',
           barmode='group',
           color_discrete_map = color_dict,
           labels = {'Value':'Environmental Impact'},
           log_y = True,
          width=1000,
          height=600,
           title = 'Cradle-to-gate analysis for 250 g item(partial product life cycle):')
fig.show()
```
